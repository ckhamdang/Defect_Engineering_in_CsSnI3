```python
import pandas as pd
import numpy as np
from scipy import stats 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_0 = pd.read_excel('feature for doping on Sn-site-I-rich.xlsx')
data_0
```

```python
data_1 = data_0.drop(columns=['Unnamed: 0'])
stat_1 = stats.zscore(data_1)
```

```python
data_1.var()
```

```python
corr_1 = stat_1.corr().round(2)
plt.figure(figsize=(14,8))
sns.heatmap(corr_1, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
#plt.savefig('corr_1.jpeg', format='jpeg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
data_2 = data_1.drop(columns=['EN', 'X', 'M', 'AR', 'CR', 'IR', 'HF (exp)'])
display (data_2)
```

```python
stat_2 = stats.zscore(data_2)
```

```python
corr_2 = stat_2.corr().round(2)
plt.figure(figsize=(14,8))
sns.heatmap(corr_2, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
#plt.savefig('corr_2.jpeg', format='jpeg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
corr_2 = stat_2.corr().round(2)
plt.figure(figsize=(14,8))
mask = np.triu(np.ones_like(corr_2))
sns.heatmap(corr_2, annot=True, mask=mask, cmap='coolwarm', vmin=-1, vmax=1)
plt.savefig('linear_corr_Irich.jpeg', format='jpeg', dpi=300, bbox_inches='tight')
plt.show()
```
